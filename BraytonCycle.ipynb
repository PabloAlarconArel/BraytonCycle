#**Ciclo Brayton con Recalentamiento Intermedio**

Ciclo_brayton.png

El ciclo Brayton con recalentamiento intermedio es una modificación del ciclo Brayton simple, que se utiliza principalmente en turbinas de gas para mejorar su rendimiento térmico y su potencia de salida. Este ciclo incluye una etapa adicional de recalentamiento del aire o gas de trabajo entre dos expansiones en turbina.

Etapas del ciclo Brayton con recalentamiento intermedio.

1. Compresión isentrópica (en el compresor).

2. Calentamiento isobárico primario (en la cámara de combustión principal).

3. Primera expansión isentrópica (en la primera turbina).

4. Recalentamiento isobárico (en una segunda cámara de combustión).

5. Segunda expansión isentrópica (en la segunda turbina).

6. Enfriamiento o escape del gas.
